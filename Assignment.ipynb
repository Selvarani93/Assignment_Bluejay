```python
df['Time Diff'] = df.groupby('Employee Name')['Time'].diff()
#calculating difference between consecutive shifts
```

```python
filtered_df = df[(df['Time Diff'] > '1:00:00') & (df['Time Diff'] < '10:00:00')]
#filtering the df
```

```python
filtered_df[['Employee Name', 'Position ID', 'Position Status']].drop_duplicates()
#Displaying
```

**THIS ANSWERS THE SECOND QUESTION, who have less than 10 hours of time between shifts but greater than 1 hour**

let's see the first occurences of each employee having this criteria.

```python
unique_names_first_occurrence_df = filtered_df.groupby('Employee Name').first().reset_index()[['Employee Name', 'Time', 'Time Diff']]
unique_names_first_occurrence_df
```

```python
df['Shift Duration'] = df['Time Out'] - df['Time']
#calculating the duration of each shifts
```

```python
long_shifts_df = df[df['Shift Duration'] > '14:00:00']
#filtering the df which are greater than 14 hours
```

```python
print(long_shifts_df[['Employee Name', 'Time', 'Position ID']])
#displaying
```
